### Explanation of Methodology

Start at origin of the Tick Fire and creating a grid of points 1 km apart that will end up creating a 15 km by 15 km grid using latitude and longitude. For converting the units I am using the convention that (the cosine in the longitude is due to the curvature of the earth but will be ignored in calculation since it only changes the outcome on the order of less than 10 meters):
\begin{equation}
latitude\; (y): 1 \;degree = 110.574 \;km \implies 1 \;km \approx 0.009044\; degrees\\
longitude\; (x): 1\; degree = 111.320cos(y) \;km \implies 1 \;km \approx 0.008983  \; degrees
\end{equation}




```python
#start values for Getty Fire
#origin src=https://www.fire.ca.gov/incidents/2019/10/28/getty-fire/
# x is longitude, y is latitude
x_g = -118.478717
y_g = 34.098191
```

```python
#creating a 15x15 km square (225 points!)
n_vals = list(range(0, 8)) + [-i for i in range(1, 8)]
#for twitter scrape geo coords have to be of the form "latitude,longitude,1km"
longs = [str(round(x_g + .009044 * n, 6)) + ",1km" for n in n_vals]
lats = [str(round(y_g + .008983 * n, 6)) + "," for n in n_vals]
```

```python
out = [lat+long for lat in lats for long in longs ]
```
